```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('simulated_error_logs.csv')
df
```

```python
df.describe()
```

```python
df['datetime_object'] = pd.to_datetime(df['Status_Ts'])
df['timestamp'] = df['datetime_object'].astype(int) / 10**9
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['timestamp'], df['Active_Errors_Count'], marker='o', linestyle='-')
plt.title('Erreurs')
plt.xlabel('Timestamp')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df_sub = df.head(25)
plt.figure(figsize=(12, 6))
plt.step(df_sub['datetime_object'], df_sub['Active_Errors_Count'], where='post')
plt.title('Erreurs')
plt.xticks([])
plt.yticks([])
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
plt.tight_layout()
plt.show()
```
